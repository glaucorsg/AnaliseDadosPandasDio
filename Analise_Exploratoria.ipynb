```python
# Importando bibliotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Criando dataframe e visualizando as primeiras cinco linhas

df = pd.read_excel("AdventureWorks.xlsx")
df.sample(10)
```

```python
# Verificando a quantidade de linhas e colunas do dataframe

df.shape
```

```python
# Verificando os tipos de dados

df.dtypes
```

```python
# Valor da receita total 

df["Valor Venda"].sum()
```

```python
# Valor do custo da venda

df["Custo da Venda"] = df["Custo Unitário"]*df["Quantidade"]
df.sample(5)
```

```python
# Cálculo do custo total de venda

round(df["Custo da Venda"].sum(),2)
```

```python
# Cálculo do lucro por venda

df["Lucro"] = df["Valor Venda"] - df["Custo da Venda"]
```

```python
# Cálculo do lucro total

round(df["Lucro"].sum(),2)
```

```python
# Tempo até o envio (em dias)

df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
df.sample(5)
```

```python
# Transformando a data de envio em um dado numérico

df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.sample(2)
```

```python
# Calculando o tempo médio de envio

round(df.groupby("Marca")["Tempo_envio"].mean(),2)
```

```python
# Verificando valores ausentes

df.isnull().sum()
```

```python
# Formantando e calculando o lucro por ano e por marca

pd.options.display.float_format = '{:20,.2f}'.format

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Resetando o index

lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Total de produtos vendidos

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False)
```

```python
# Gráfico para o total de vendas por produto

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True).plot.barh(title = "Total de produtos vendidos")
plt.xlabel("Total de vendas")
plt.ylabel("Produto");
```

```python
# Gráfico para o lucro por ano

df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().sort_values(ascending = False).plot.bar(title = "Lucro por ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
# Vendas de 2009

vendas_2009 = df[df["Data Venda"].dt.year == 2009]
vendas_2009.sample(2)
```

```python
# Lucro mensal para 2009

vendas_2009.groupby(vendas_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro mensal em 2009", marker = "o")
plt.xlabel("Mês")
plt.ylabel("Lucro (R$)");
```

```python
# Lucro por marca

vendas_2009.groupby("Marca")["Lucro"].sum().plot.bar(title = "Lucro por marca em 2009")
plt.xlabel("Marca")
plt.ylabel("Lucro")
;
```

```python
# Lucro por classe

vendas_2009.groupby("Classe")["Lucro"].sum().plot.bar(title = "Lucro por marca em 2009")
plt.xlabel("Classe")
plt.ylabel("Lucro");
```
